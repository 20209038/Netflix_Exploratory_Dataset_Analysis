```python
data_1 = data.copy()
```

```python
plt.figure(figsize=(12,6))
plt.title("Percentation of Netflix Titles that are either Movies or TV Shows")
g = plt.pie(data_1.Category.value_counts(),explode=(0.025,0.025),labels=data_1.Category.value_counts().index, colors=['red','yellow'],textprops={"fontsize":15},autopct='%1.1f%%',shadow=True, startangle=180)
plt.legend(loc=3)
plt.show()
```

```python
# dealing will null values
data_1['Director'].fillna('No Director', inplace=True)
data_1['Cast'].fillna('No Cast', inplace=True)
data_1['Country'].fillna('Country Unavailable', inplace=True)
data_1.dropna(subset=['Release_Date','Rating'],inplace=True)
```

```python

filtered_genres = data_1.set_index('Title').Type.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);

plt.figure(figsize=(7,9))
g = sns.countplot(y = filtered_genres, order=filtered_genres.value_counts().index[:20])
plt.title('Top 20 Genres on Netflix')
plt.xlabel('Titles')
plt.ylabel('Genres')
plt.show()
```

```python
#Who are the top 10 Movies actors on Netflix based on number of titles?

filtered_cast = data_1[data_1.Cast != 'No Cast'].set_index('Title').Cast.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
sns.countplot(y = filtered_cast, order=filtered_cast.value_counts().index[:10], palette='rocket')
plt.show()
```

```python
filtered_directors = data_1[data_1.Director != 'No Director'].set_index('Title').Director.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
sns.countplot(y = filtered_directors, order=filtered_directors.value_counts().index[:10], palette='mako')
plt.show()
```
